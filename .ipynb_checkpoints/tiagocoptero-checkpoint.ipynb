```python
%matplotlib inline
import os, glob
import matplotlib.image as img
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
mypath='/home/fer/Pictures/stolenimages/'
lista = []
for infile in sorted(glob.glob( os.path.join(mypath, '*.tiff')) ):
    lista.append(infile)
```

```python
print(len(lista)) #number of images
```

```python
time = np.linspace(0,74./30,74) #30 frames per second
```

```python
plt.imshow(img.imread(lista[37])) #sneak peeking before clipping
plt.grid()
```

```python
cubo0,cubo1,cubo2 = [],[],[]
for i in range(len(lista)):
    #print(lista[i])
    im = img.imread(lista[i])
    im0 = im[170:670,90:600,0] #clipping every image layer
    im1 = im[170:670,90:600,1]
    im2 = im[170:670,90:600,2]
    val0 = np.sum(im0)         #summation of brightness
    val1 = np.sum(im1)
    val2 = np.sum(im2)
    cubo0.append(val0)         #append to a list
    cubo1.append(val1)
    cubo2.append(val2)
```

```python
plt.imshow(im0) #sneak peeking after clipping
```

```python
curva0 = signal.detrend(cubo0)
curva1 = signal.detrend(cubo1)
curva2 = signal.detrend(cubo2)
```

```python
c0 = (curva0-np.min(curva0))/np.max(curva0) 
c1 = (curva0-np.min(curva1))/np.max(curva1)
c2 = (curva0-np.min(curva2))/np.max(curva2)
```

```python
m0 = np.mean([c0[0],c0[73]])-c0[37] # a kind of quality index of (transit curve + detrend process) by layer
m1 = np.mean([c1[0],c1[73]])-c1[37]
m2 = np.mean([c2[0],c2[73]])-c2[37]
```

```python
w0=1/m0/(1/m0+1/m1+1/m2) # building weighting indices
w1=1/m1/(1/m0+1/m1+1/m2)
w2=1/m2/(1/m0+1/m1+1/m2)
```

```python
transit = np.mean([c0*w0,c1*w1,c2*w2],axis=0)
transit = (transit -np.min(transit))/np.max(transit)-1.0
error = np.std([c0*w0,c1*w1,c2*w2],axis=0)
```

```python
plt.figure(figsize=(8,6))
plt.errorbar(time,transit,yerr=error,fmt='o',color='r')
plt.plot(time,transit,'k--')
plt.xlim(-0.01,2.49)
plt.xlabel('time (s)')
plt.ylabel('brightness')
plt.savefig('tiagocoptero.png')
```
